```python
from wrapper  import Kanbanize
import xml.etree.ElementTree as ET
import pandas as pd

import numpy as np
import sys


empresa = ""
apikey = ""

if empresa == "" or apikey =="":
    print("")
    sys.exit("Error message:ingrese la empresa y la apikey")
    
k = Kanbanize(empresa, apikey)

r1 = k.get_all_tasks(37,format='dict')
df = pd.DataFrame(r1)

#Ordenar Columnas 
respuesta = df.reindex(['taskid','position','type','assignee','title','description','subtasks','subtaskscomplete','color','priority','size','deadline','deadlineoriginalformat','extlink','tags','columnid','laneid','leadtime','blocked','blockedreason','columnname','lanename','logedtime','boardparent','archivedat','reporter','createdat','workflow_name','workflow,workflow_id','columnpath','linkschild','linksparent','linksrelated','linkspredecessor','linkssuccessor','updatedat'], axis=1)

r1 = respuesta.groupby('columnname')['taskid'].agg(['count'])
r3 = respuesta.groupby('assignee')['columnname'].agg(['count'])

r4 = respuesta.groupby('deadlineoriginalformat')['taskid'].agg(['count'])
r5 = respuesta.groupby('deadline')['leadtime'].agg(['mean'])

#Agrupacion por Asignado y Columna
grp = respuesta.groupby(['assignee',  'columnpath', 'columnname'])['leadtime'].count()

#Genera archivo csv
respuesta.to_csv (r'./export_dataframe.csv', index = None, header=True) 
grp.to_csv (r'./group_dataframe.csv', index = True, header=True)

print(r1)
print(r3)
print(r4)
print(r5)

print(grp)
```
